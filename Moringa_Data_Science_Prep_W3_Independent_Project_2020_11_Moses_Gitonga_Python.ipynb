```
# Upload the CDR_description dataset and preview the first 3 lines

df_cells_geo =pd.read_excel("/content/CDR_description.xlsx")
df_cells_geo.head(3)
```

```
# Upload the Telkom May 7th dataset
# The headers in the May 7th, 8th and 9th data are different so once I upload I change the column headers 
# so that they are standard.

df_may_07 = pd.read_csv("/content/Telcom_dataset.csv")
df_may_07

df_may_07.columns = ['Product', 'Value', 'Date_Time', 'Cell_On_Site', 'Caller Details ', 'Receiver Details', 'Caller Country',
              'Receiver Country', 'Cell_ID', 'Site_ID']
df_may_07
```

```
# Upload the Telkom May 8th dataset
# The headers in the May 7th, 8th and 9th data are different so once I upload I change the column headers 
# so that they are standard.

df_may_08 = pd.read_csv("/content/Telcom_dataset2.csv")

df_may_08.columns = ['Product', 'Value', 'Date_Time', 'Cell_On_Site', 'Caller Details ', 'Receiver Details', 
                     'Caller Country', 'Receiver Country', 'Cell_ID', 'Site_ID']
df_may_08
```

```
# Upload the Telkom May 9th dataset
# The headers in the May 7th, 8th and 9th data are different so once I upload I change the column headers 
# so that they are standard.

df_may_09 = pd.read_csv("/content/Telcom_dataset3.csv")

df_may_09.columns = ['Product', 'Value', 'Date_Time', 'Cell_On_Site', 'Caller Details ', 'Receiver Details', 
                     'Caller Country', 'Receiver Country', 'Cell_ID', 'Site_ID']

df_may_09
```

```
# Merging dataframes for the three days

may_all = pd.concat([df_may_07, df_may_08, df_may_09], axis=0, ignore_index=True)

# display the first five records of the merged dataframe
may_all
```

```
# Get summary statistics for the data
may_all.describe()
```

```
# Delete transactional data from the may_all dataset that does not have values in the Site and/or cell IDs
may_all_updated = may_all.dropna (thresh=10)
may_all_updated
```

```
# Merge the may_all_updated database above with the cells_geo dataset
df1_cleaned = pd.merge(may_all_updated, df_cells_geo, how='left', on='Site_ID').dropna()

df1_cleaned
```

```
# drop columns that are not necessary for the analysis

df2_cleaned = df1_cleaned.drop(["Value","Receiver Details", "Caller Details ", "Caller Country", "Receiver Country", "Cell_ID_x", "Site_ID", "No", "Longitude", "Latitude",
                 "Cell_ID_y"], axis=1)
df2_cleaned
```

```
# drop duplicates

df3_cleaned = df2_cleaned.drop_duplicates()

df3_cleaned
```

```
# Now that the data has been cleaned, analyze the data.
# Start by analyzing which was the most active city

# Create a variable that will be used to calculate the most active city
city_ranked = df3_cleaned['City'].groupby(df3_cleaned['City'])

# Thereafter, display the length of the cities that have been sorted

len(city_ranked) # Showing that there are 396 cities in total that the telco has coverage in
```

```
# Next, check the largest cities by telco traffic

city_ranked.count().nlargest(10)
```

```
# Next, check the smallest cities by telco traffic

city_ranked.count().nsmallest(10)
```

```
# Check the peak hours

city_ranked = df3_cleaned['City'].groupby(df3_cleaned['Date_Time'])
city_ranked.count().nlargest(25)
```

```
# Check the off-peak hours

city_ranked = df3_cleaned['City'].groupby(df3_cleaned['Date_Time'])
city_ranked.count().nsmallest(25)
```
